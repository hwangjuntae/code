```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
```

## 1) Load Image

```python
img = cv2.imread('painting.tif', cv2.IMREAD_GRAYSCALE) # 이미지 파일 읽기

cv2.imshow('image', img)      
cv2.waitKey(0)                
cv2.destroyAllWindows()       
```

## 2) Define Function

**요구사항**

- 1) Resizing, translation, rotation, shear 함수를 순서대로 OpenCV의 warpAffine 함수를 이용하여 구현하세요.
- 2) Resizing 함수는 결과 이미지를 1.8배 확대하도록 구현하세요.
- 3) Translation 함수는 결과 이미지를 가로와 세로 방향으로 각각 40픽셀, 90픽셀 만큼 이동하도록 구현하세요.
- 4) Rotation 함수는 아래 matrix를 참고하여 결과 이미지를 반시계방향으로 30도 rotate하도록 구현하세요.(회전축은 이미지의 중심) 
- 5) Shear 함수는 결과 이미지를 가로 0.2배, 세로 0.1배로 shearing 하는 이미지를 만들도록 구현하세요.

<div>

</div>

```python
def my_resizing(image):
    matrix = np.array([[1.8, 0, 0], [0, 1.8, 0]], dtype=np.float32)
    result = cv2.warpAffine(image, matrix, (image.shape[1], image.shape[0]))

    return result
```

```python
def my_translation(image):
    matrix = np.array([[1, 0, 40], [0, 1, 90]], dtype=np.float32)
    result = cv2.warpAffine(image, matrix, (image.shape[1], image.shape[0]))

    return result
```

```python
def my_rotation(image, angle):
    center_x = image.shape[1] / 2
    center_y = image.shape[0] / 2
    matrix = cv2.getRotationMatrix2D((center_x, center_y), angle, 1)
    result = cv2.warpAffine(image, matrix, (image.shape[1], image.shape[0]))

    return result
```

```python
def my_shear_x(image):
    matrix = np.array([[1, 0.2, 0], [0, 1, 0]], dtype=np.float32)
    result = cv2.warpAffine(image, matrix, (image.shape[1], image.shape[0]))

    return result

def my_shear_y(image):
    matrix = np.array([[1, 0, 0], [0.1, 1, 0]], dtype=np.float32)
    result = cv2.warpAffine(image, matrix, (image.shape[1], image.shape[0]))

    return result
```

```python
resized = my_resizing(img)
translation = my_translation(img)
rotation = my_rotation(img, 30)
shearing = my_shear_y(my_shear_x(img))
```

## 3) Plot Image

```python
plt.figure(figsize = (20, 12))
images = {'Original':img, 'Image Enlaged':resized, 'Image Translated':translation,
         'Image Rotated':rotation, 'Image Shearing':shearing}

for i, (k, v) in enumerate(images.items()):
    plt.subplot(2,3,i+1)
    plt.title(k)
    plt.imshow(cv2.cvtColor(v, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    
plt.show()
```

## 2) Define Function

**요구사항**

- 1) 아래 코드들을 실행하면 set points 윈도우에 마우스로 꼭짓점 4개를 지정할 수 있습니다.
- 2) 카드 이미지(img)에서 spade 10의 4개 꼭짓점을 클릭하세요.
- 3) OpenCV의 Perspective Transform 함수를 사용하여 변환 행렬을 구하고 perspective transform을 적용하세요.

```python
# find_position 함수는 mouse로 받은 points 들을 입력으로 받아 각 points들의 위치를 구해서 반환합니다.
# output_size 함수를 통해 output size (width, height)를 정해 반환합니다.
# 아래 함수는 바꾸지 마세요.

def find_position(points):
    sum_xy = np.sum(points, axis=1)
    diff_xy = np.diff(points, axis=1)
    
    topLeft = points[sum_xy.argmin()]
    bottomRight = points[sum_xy.argmax()]
    topRight = points[diff_xy.argmin()]
    bottomLeft = points[diff_xy.argmax()]
    
    return topLeft, topRight, bottomRight, bottomLeft

def output_size(topLeft, topRight, bottomRight, bottomLeft):
    w1, w2 = bottomRight[0] - bottomLeft[0], topRight[0] - topLeft[0]
    h1, h2 = bottomRight[1] - topRight[1], bottomLeft[1] - topLeft[1] 
    
    return int(max(w1, w2)), int(max(h1, h2))
```

```python
# on_mouse 함수는 마우스로 정한 좌표 4개를 받아 변환 전 좌표(src)와 변환 후 좌표(dst)을 정합니다. 
# 아래 함수는 바꾸지 마세요.

def on_mouse(event, x, y, flags, param):
    global count, src, dst, width, height
    
    if event == cv2.EVENT_LBUTTONDOWN:       
        cv2.circle(draw_points, (x, y), 10, (0,0,255), 3)
        cv2.imshow('set points', draw_points)

        points[count] = (x, y)
        count += 1
        if count == 4:
            topL, topR, bottomR, bottomL = find_position(points)
            
            src = np.float32([topL, topR, bottomR, bottomL])
            cv2.polylines(draw_points, [np.array(src,np.int32)], True, (0,255,0), 5)
            
            width, height = output_size(topL, topR, bottomR, bottomL)

            dst = np.float32([[0, 0], [width-1, 0], [width-1, height-1], [0, height-1]])
            cv2.destroyAllWindows()
```

```python
# Perspective Transform을 실행할 Source를 mouse를 이용하여 정합니다
# 아래 코드를 바꾸지 마세요.

count = 0
points = np.zeros((4, 2))
draw_points = img.copy()

cv2.imshow('set points', img)
cv2.setMouseCallback('set points', on_mouse)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Perspective Transform을 실행한 Destination을 보여줍니다

matrix = cv2.getPerspectiveTransform(src, dst)
result = cv2.warpPerspective(img, matrix, (width, height))
cv2.imshow('result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 3) Plot Image

```python
plt.figure(figsize = (15, 15))
images = {'Original':img, 'Points':draw_points, 'Perspective Transform':result}

for i, (k, v) in enumerate(images.items()):
    plt.subplot(2,2,i+1)
    plt.title(k)
    plt.imshow(cv2.cvtColor(v, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    
plt.show()
```

## 4) Discussion

**1) Affine Transform과 Perspective Transform의 차이에 대해 설명하세요.**


Affine Transform과 Perspective Transform의 차이는 변환에 사용되는 포인트의 개수와 변환의 종류이다.

Affine Transform은 평면 상의 점들의 선형 변환을 수행하는데, 이를 통해 이미지나 객체의 크기 조절, 회전, shear, 이동 등의 변환을 수행한다. 

Affine Transform은 이미지나 객체의 형태 유지와 변환된 결과의 평행성 유지에 적합하다. 

Perspective Transform은 3D 공간을 2D 평면에 투영하는 변환이다. 

Perspective Transform은 원근감을 통해 Affine Transfrom의 기능과 더불어 원근 효과를 구현할 수 있다. 

Perspective Transform은 카메라 투영, 3D 장면의 2D 투영 등에서 사용된다.

결론적으로, Affine Transform은 선형적 변환에 사용되며, 이미지의 형태 유지를 중요시한다. 
반면, Perspective Transform은 원근감을 포함한 비선형적인 변환에 사용되며, 원근 효과를 구현할 수 있다.

**2) Inverse mapping 방법을 forward mapping과 비교하여 설명하세요.**

Forward mapping은 원본 이미지에서 대상 이미지로 픽셀을 직접 매핑하는 방법이다. 

각 원본 이미지의 픽셀 위치에 대해 변환 행렬을 적용하여 대상 이미지의 해당 위치로 픽셀을 이동시킨다. 
하지만 forward mapping은 대상 이미지에 대응되지 않는 원본 이미지의 픽셀이나 변환 후의 이미지 영역을 채우지 못할 수 있다.

Inverse mapping은 대상 이미지에서 원본 이미지로 픽셀을 매핑하는 방법이다. 

대상 이미지의 각 픽셀 위치에 대해 역변환 행렬을 적용하여 원본 이미지의 해당 위치로 픽셀을 이동시킨다. 
Inverse mapping은 대상 이미지에서 원본 이미지로의 대응 관계를 구하는 데 사용된다. 
이 방법은 forward mapping과는 달리 모든 대상 이미지의 픽셀이 변환에 참여하기에, 변환 후 이미지 영역을 빠짐없이 채운다.

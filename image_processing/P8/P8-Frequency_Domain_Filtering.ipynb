## Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
```

## 1) Load Image

```python
ori = cv2.imread('original.tif', cv2.IMREAD_GRAYSCALE)
noisy = cv2.imread('noisy_boulder.tif', cv2.IMREAD_GRAYSCALE)
```

## 2) Define Function

**요구사항**

- 1) 입력 이미지(f)의 가로와 세로 2배 길이로 zero padding 합니다 (이부분은 주어집니다)
- 2) Padding 한 이미지를 FFT 후 FFT shift 시키세요 (e.g. np.fft.fft2, np.fft.fftshift)
- 3) 강의 내용을 참고하여 Ideal, Gaussian, Butterworth Low Pass Filter를 구현하여 transfer function을 구하세요
- 4) 앞서 구한 2)와 3)의 결과를 곱하세요
- 5) 앞서 구한 4)의 결과를 Inverse FFT shift 후 Inverse FFT 시키세요
- 6) 앞서 구한 5)의 결과를 원본 이미지 크기로 crop 후 실수값을 구하세요

```python
# Low pass filter 함수는 input image, cutoff frequency를 입력으로 받아 
# filtering한 결과, 이미의 주파수 영역에서의 값, transfer function을 반환합니다

def lowpass_filter(f, D_zero, n=1, filter_type='Ideal'):
    f_padded = np.pad(f, ((0, f.shape[0]), (0, f.shape[1])), 'constant', constant_values=0) # 이미지 패딩

    F = np.fft.fftshift(np.fft.fft2(f_padded)) # FFT & FFT shift

    # 주파수 도메인에서 좌표를 구하기 위한 코드
    P, Q = f_padded.shape
    fy, fx = np.linspace(0, P - 1, P), np.linspace(0, Q - 1, Q)
    [v, u] = np.meshgrid(fx, fy)

    D = np.sqrt((u - P/2)**2 + (v - Q/2)**2) # 주파수 도메인에서 center 위치에서 거리

    if filter_type == 'Ideal':
        H = np.where(D <= D_zero, 1, 0)
    elif filter_type == 'Gaussian':
        H = np.exp(-(D**2) / (2 * (D_zero**2)))
    elif filter_type == 'Butterworth':
        H = 1 / (1 + (D / D_zero)**(2 * n))
    else:
        print("Wrong Filter Type")

    G = F * H
    g_padded = np.fft.ifftshift(G) # iFFT shift
    g = np.fft.ifft2(g_padded).real # iFFT & 실수값 변환
    image = g[:f.shape[0], :f.shape[1]] # 이미지 크롭
    image = np.clip(image, 0, 255) # 값 범위 조정

    return image.astype('uint8'), F, H
```

```python
# Cutoff Frequency (D_zero)에 적절한 값을 넣어 denoising 하세요
# 3개의 Low Pass Filter에 같은 값의 Cutoff Frequency 값을 넣으세요

D_zero = 50

ILPF, F, H_i = lowpass_filter(noisy, D_zero, filter_type='Ideal')
GLPF, _, H_g = lowpass_filter(noisy, D_zero, filter_type='Gaussian')
BLPF, _, H_b = lowpass_filter(noisy, D_zero, n=2.5, filter_type='Butterworth')
```

## 3) Plot Image

```python
fig = plt.figure(figsize = (20, 35))
images = {'Original Image':ori, 'Noisy Image':noisy, 'Ideal Low Pass Filter':ILPF,
         'Gaussian Low Pass Filter':GLPF, 'Buterworth Low Pass Filter':BLPF}

padded = np.pad(ori, ((0, ori.shape[0]), (0, ori.shape[1])), 'constant', constant_values = 0)
ori_freq = 10*np.log(np.abs(np.fft.fftshift(np.fft.fft2(padded))))
frequency_domain = {'Original Image at Frequency Domain':ori_freq, 
                    'Noisy Image at Frequency Domain':10*np.log(np.abs(F)), 
                    'ILPF Transfer Function':H_i, 'GLPF Transfer Function':H_g,
                    'BLPF Transfer Function':H_b}

for i, ((k, v),(f_k, f_v)) in enumerate(zip(images.items(), frequency_domain.items())):
    plt.subplot(5,2,2*i+1)
    plt.title(k)
    plt.imshow(cv2.cvtColor(v, cv2.COLOR_GRAY2RGB))
    plt.axis("off")
    
    plt.subplot(5,2,2*i+2)
    plt.title(f_k)
    plt.imshow(f_v, cmap='gray')
    plt.axis("off")
    
plt.show()
```

## 1) Load Image

```python
img = cv2.imread('char.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imshow('image', img)      
cv2.waitKey(0)                
cv2.destroyAllWindows()   
```

## 2) Define Function

**요구사항**

- 1) 입력 이미지(f)의 가로와 세로 2배 길이로 zero padding합니다
- 2) Padding 한 이미지를 FFT 후 FFT shift 시키세요 (e.g. np.fft.fft2, np.fft.fftshift)
- 3) 강의 내용을 참고하여 Ideal, Gaussian, Butterworth High Pass Filter를 구현하여 transfer function을 구하세요
- 4) 앞서 구한 2)와 3)의 결과를 곱하세요
- 5) 앞서 구한 4)의 결과를 Inverse FFT shift 후 Inverse FFT 시키세요
- 6) 앞서 구한 5)의 결과를 원본 이미지 크기로 crop 후 실수값을 구하세요

```python
import numpy as np
import cv2

def highpass_filter(f, D_zero, n=1, filter_type='Ideal'):
    f_padded = np.pad(f, ((0, f.shape[0]), (0, f.shape[1])), 'constant', constant_values=0) # 이미지 패딩

    F = np.fft.fftshift(np.fft.fft2(f_padded)) # FFT & FFT shift

    # 주파수 도메인에서 좌표를 구하기 위한 코드
    P, Q = f_padded.shape
    fy, fx = np.linspace(0, P - 1, P), np.linspace(0, Q - 1, Q)
    [v, u] = np.meshgrid(fx, fy)

    D = np.sqrt((u - P/2)**2 + (v - Q/2)**2) # 주파수 도메인에서 center 위치에서 거리

    if filter_type == 'Ideal':
        H = np.where(D > D_zero, 1, 0)
    elif filter_type == 'Gaussian':
        H = 1 - np.exp(-(D**2) / (2 * (D_zero**2)))
    elif filter_type == 'Butterworth':
        H = 1 / (1 + (D_zero / D)**(2 * n))
    else:
        print("Wrong Filter Type")

    G = F * H
    g_padded = np.fft.ifftshift(G) # iFFT shift
    g = np.fft.ifft2(g_padded).real # iFFT & 실수값 변환
    image = g[:f.shape[0]//2, :f.shape[1]//2] # 이미지 크롭
    image = np.clip(image, 0, 255) # 값 범위 조정

    return image.astype('uint8'), F, H
```

```python
# 아래 코드는 수정하지 마세요
IHPF, F, H_i = highpass_filter(img,80,filter_type='Ideal')
GHPF, _, H_g = highpass_filter(img,80,filter_type='Gaussian')
BHPF, _, H_b = highpass_filter(img,80,filter_type='Butterworth')
```

## 3) Plot Image

```python
fig = plt.figure(figsize = (20, 30))
images = {'Original Image':img, 'Ideal High Pass Filter': np.clip(IHPF,0,255), 
          'Gaussian High Pass Filter':GHPF, 'Butterworth High Pass Filter':BHPF}

frequency_domain = {'Image at Frequency Domain':10*np.log(np.abs(F)),
                    'IHPF Transfer Function':H_i, 
                    'GHPF Transfer Function':H_g,
                    'BHPF Transfer Function':H_b}

for i, ((k, v),(f_k, f_v)) in enumerate(zip(images.items(), frequency_domain.items())):
    plt.subplot(4,2,2*i+1)
    plt.title(k)
    plt.imshow(cv2.cvtColor(v, cv2.COLOR_GRAY2RGB))
    plt.axis("off")
    
    plt.subplot(4,2,2*i+2)
    plt.title(f_k)
    plt.imshow(f_v, cmap='gray')
    plt.axis("off")
    
plt.show()
```

## 4) Discussion

**1) Frequency domain에서 filtering 할 때 Padding과 Centering을 하는 이유에 대해 설명하세요.**

1. 주파수 정보의 손실 방지
 입력 이미지가 작을 경우 주파수 도메인에서 나타나는 정보가 제한될 수 있다. Padding을 통해 입력 이미지를 확장하여 보다 더 정확한 주파수 도메인의 결과를 얻을 수 있다.

2. 주파수 도메인에서의 중심 이동 보정
 영상의 주파수 영역에서 중심은 일반적으로 저주파수 성분을 나타낸다. 따라서, 주파수를 이동하여 주파수 도메인 중심을 이미지의 중앙으로 이동시킨다. 중심을 옮기는 작업은 주파수 도메인에서 필터링을 수행할 때 필터의 적용 위치를 조절하기 위해 필요하다.

**2) Zero-padding을 하지 않을 경우 어떤 결과가 나오는 지 돌려보고 분석하세요.**


Zero-padding을 하지 않으면 주파수 도메인에서 나타나는 결과의 해상도를 제한하게 된다.

구체적으로, 주파수 정보가 부족하여 고주파 성분이 표현되지 않거나 정확한 위치에서 나타나지 않을 수 있다. 또한, 스펙트럼이 희석된다. 주파수 도메인에서의 픽셀 간격이 증가하여 스펙트럼이 뭉게져서 세부 특징이 흐려진다. 

**3) Frequency domain에서 filtering 후 Spatial domain에서 실수(Real)값만 뽑아서 쓰는 이유에 대해 설명하세요.**


1. 공간 도메인의 특성 유지
실수 값을 사용하면 이미지의 밝기, 색상 및 명암 등의 공간적인 특성을 보존 가능하다.

2. 변환할 때의 편리성
실수 값만을 사용해서 역변환을 하면, 원본 이미지의 공간 도메인으로 효과적으로 복원할 수 있다.

**4) Ideal Low Pass Filter에서 cutoff frequency가 클 때와 작을 때의 장단점에 대해 설명하세요.**

1. cutoff frequency가 클 경우

장점
세부적인 고주파 성분을 보존할 수 있기에, 이미지의 선명도가 유지된다.

단점
노이즈도 통과할 수 있기에, 노이즈가 이미지에 영향을 줄 수 있다.
넓은 주파수 범위의 영향을 받게 된다.

2. cutoff frequency가 작을 경우

장점
이미지의 부드러운 성분이 강조된다.

단점
세부적인 고주파 성분이 손실되기에, 이미지의 선명도가 감소한다.
부드러운 성분이 강조되므로, 이미지의 경계가 흐려진다.

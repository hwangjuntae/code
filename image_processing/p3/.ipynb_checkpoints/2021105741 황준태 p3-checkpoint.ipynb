### 직접 구현할 코드

```python
def my_gaussian_filter(image, kernel_size, sigma, visualization_toggle = False):
    # my_gaussian_filter 함수는 input 이미지를 받아서 kernel size와 sigma 를 정하여 가우시안 커널을 생성한 후
    # 이미지에 filter를 적용한(opencv filter2D 함수로 filtering 한) 이미지를 return 하는 함수입니다.
    # Visualization toggle을 True로 하면 생성한 가우시안 커널을 볼 수 있습니다.

    # np.linspace 사용해서 [x, y] 좌표 생성 과정 e.g. t = [0, 1, 2, 3, 4]
    t = np.linspace(-(kernel_size-1)/2., (kernel_size-1)/2., kernel_size)
    # np.meshgrid 사용해서 [x, y] 좌표 생성 e.g. X=[-2, -1, 0 , 1, 2], Y = [-2, -1, 0, 1, 2]^t (transposed)
    [X, Y] = np.meshgrid(t, t)
    # Gaussian Filter 공식 구현
    formula = np.exp(-(np.square(X) + np.square(Y)) / (2*np.square(sigma)))
    # K값 구하기
    K = 1 / np.sum(formula)
    # K값 x formula 해서 Gaussian kernel 구하기
    gaussian_kernel = K * formula

    if visualization_toggle: # visualization toggle을 사용하면 kernel 확인 가능
        fig = plt.figure(figsize = (10, 10))
        ax = plt.axes(projection='3d')
        ax.plot_surface(X, Y, gaussian_kernel, cmap='viridis')
        ax.set_title('Gaussian kernel')

    result = cv2.filter2D(image, -1, gaussian_kernel, borderType=cv2.BORDER_CONSTANT)
    return result
```

### OpenCV를 사용한 코드

```python
gaussian_filter_opencv_9_1 = cv2.GaussianBlur(img, (9, 9), 1)
gaussian_filter_opencv_25_2 = cv2.GaussianBlur(img, (25, 25), 2)
bilateral_filtered = cv2.bilateralFilter(img, 5, 75, 75)
```

## 3) Plot Image

### 직접 구현한 Gaussian Filter와 OpenCV Gaussian Filter 비교

```python
fig = plt.figure(figsize = (15, 10)) 
rows = 2
cols = 3
  
ax1 = fig.add_subplot(rows, cols, 1)
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))
ax1.set_title('Original')
ax1.axis("off")
 
ax2 = fig.add_subplot(rows, cols, 2)
ax2.imshow(cv2.cvtColor(my_gaussian_filter(img, 9, 1), cv2.COLOR_GRAY2RGB))
ax2.set_title('(Kernel: 9x9, sigma: 1)')
ax2.axis("off")

ax3 = fig.add_subplot(rows, cols, 3)
ax3.imshow(cv2.cvtColor(gaussian_filter_opencv_9_1, cv2.COLOR_GRAY2RGB))
ax3.set_title('(Kernel: 9x9, sigma: 1 - OpenCV)')
ax3.axis("off")

ax4 = fig.add_subplot(rows, cols, 5)
ax4.imshow(cv2.cvtColor(my_gaussian_filter(img, 25, 2), cv2.COLOR_GRAY2RGB))
ax4.set_title('(Kernel: 25x25, sigma: 2)')
ax4.axis("off")

ax5 = fig.add_subplot(rows, cols, 6)
ax5.imshow(cv2.cvtColor(gaussian_filter_opencv_25_2, cv2.COLOR_GRAY2RGB))
ax5.set_title('(Kernel: 25x25, sigma: 2 - OpenCV)')
ax5.axis("off")

plt.show()
```

## 1) Load Image

```python
img = cv2.imread('hero2.jpeg', cv2.IMREAD_GRAYSCALE)
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 2) Define Function

**요구사항**
- 1) Unsharp Masking을 수업 내용 바탕으로 직접 구현하세요
- 2) 자신이 생각하는 가장 보기 좋게 샤프닝 된 결과가 출력되도록 parameter(Kernel Size, sigma, k)로 세팅하세요

```python
def unsharp_masking(image, k):
    # step1
    # 자신이 직접 구현한 Gaussian Filter 혹은 OpenCV의 Gaussian Filter로 blurred된 이미지 생성
    blurred_image = my_gaussian_filter(img, 9, 1)

    # step2
    # 원본 영상에서 blurred된 이미지를 빼세요
    mask = np.array(img, dtype=np.float32) - np.array(blurred_image, dtype=np.float32) 
    mask = np.clip(mask, 0, 255) # np.clip을 사용하여 0~255 범위로 맞추기
    mask = mask.astype('uint8') # float32 -> uint8 형변환

    # step3
    sharped_image = np.array(image, dtype=np.float32) + k*mask
    sharped_image = np.clip(sharped_image, 0, 255) # np.clip을 사용하여 0~255 범위로 맞추기
    result = sharped_image.astype('uint8') # 이미지로 보이기 위해 float64 -> uint8 형변환
    return result
```

## 3) Plot Image

```python
fig = plt.figure(figsize=(20, 20))
rows = 1
cols = 3

ax1 = fig.add_subplot(rows, cols, 1)
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))
ax1.set_title('Original')
ax1.axis("off")

ax2 = fig.add_subplot(rows, cols, 2)
blurred_img = my_gaussian_filter(img, 9, 1)
ax2.imshow(cv2.cvtColor(blurred_img, cv2.COLOR_GRAY2RGB))
ax2.set_title('Blurred image')
ax2.axis("off")

ax3 = fig.add_subplot(rows, cols, 3)
sharpened_img = unsharp_masking(img, k=4)
ax3.imshow(cv2.cvtColor(sharpened_img, cv2.COLOR_GRAY2RGB))
ax3.set_title('Unsharp result')
ax3.axis("off")

plt.show()
```

## 4) Discussion

**1) Gaussian Filter와 Bilateral Filter의 차이에 대해 설명하세요.**

1. 가우시안 필터는 이미지를 흐리게 하고 노이즈를 줄이는 선형 필터이다. 중앙 픽셀을 강조하고, 중앙 픽셀과의 거리에 따라 픽셀을 억제한다. 또한, 등방성 필터이다.
2. Bilateral(양방향) 필터는 가장자리를 유지하면서 이미지를 매끄럽게 만드는 비선형 필터이다. 가우스 커널을 사용하여 이미지를 흐리게 처리하며, 이 때 인접 픽셀 간의 차이도 고려한다. 양방향 필터는 중앙 픽셀로부터의 거리와 픽셀값을 고려하여 다른 가중치를 부여한다. 또한, 비등방성 필터이다.

**2) Unsharp masking의 원리에 대해 설명하세요.**

1. 이미지 블러를 통해 노이즈를 제거한다.
2. 그 후, 이미지의 가장자리와 세부 사항을 강조하는 마스크를 만든다.
3. 원본 이미지에 마스크를 추가하여 선명도를 높인다.

이미지 블러로 가장자리와 세부 사항을 강조하는 마스크를 만들어, 원본 이미지에 추가하여 선명도를 높인다.

**3) Gaussian Filter와 Median Filter를 잡음제거 관점에서 비교하세요.**

가우시안 필터는 이웃하는 픽셀 값을 평균화하여 이미지를 부드럽게 한다. 하지만, 주변 픽셀과 값이 매우 다른 픽셀의 경우에는 비교적 덜 효과적이다.


중간값 필터는 인접한 픽셀 값을 해당 픽셀의 중간값으로 대체하여 작동한다. 중간값 필터의 경우 노이즈 외의 이미지 영향을 주지 않고 노이즈 픽셀을 식별하고 제거할 수 있으므로 주변 픽셀과 값이 매우 다른 픽셀의 노이즈도 제거할 수 있다. 그러나 인접 픽셀의 값을 고려하지 않기에 가우시안 노이즈를 줄이는 데는 덜 효과적이다.
